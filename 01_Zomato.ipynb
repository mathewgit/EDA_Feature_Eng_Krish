```python
df_country=pd.read_excel('/content/drive/MyDrive/Data Science/Krish Niak EDA and Feature Engineering/01/Zomatodataset/Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
final_df
```

```python
#Find the datatypes
final_df.dtypes
```

```python
#Find how many records are there
final_df.Country.value_counts()
```

```python
#Find how many records are there
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
#Create a Piechart
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')
```

Zomatos major transaction from India then USA and then UK

```python
final_df.columns
```

```python
#Grouping By
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
#Grouping By
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
## Observation
1.When rating is between 4.5 to 4.9 -> Excellent
2.3.4 to 4 Very good
3. Good
4.Average
```

```python

ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)

sns.barplot(x="Aggregate rating",y='Rating Count',data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y='Rating Count',hue='Rating color', data=ratings,palette=['white','red','orange','yellow','green'])
```
