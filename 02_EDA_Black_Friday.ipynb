A retail company “ABC Private Limited” wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
# Train Data Set https://www.kaggle.com/datasets/sdolezel/black-friday?resource=download&select=train.csv
# Test Data Set  https://www.kaggle.com/datasets/sdolezel/black-friday?resource=download&select=test.csv
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import the train dataset
df_train=pd.read_csv('/content/drive/MyDrive/Data Science/Krish Niak EDA and Feature Engineering/02/Dataset/train.csv/train.csv')
df_train.head()
```

```python
##  import the test data
df_test=pd.read_csv('/content/drive/MyDrive/Data Science/Krish Niak EDA and Feature Engineering/02/Dataset/test.csv/test.csv')
df_test.head()
```

```python
## Append both train and test data
df=df_train.append(df_test)
df.head()
```

```python
##Basic 
df.info()
```

```python
df.describe()
```

```python
# Delete user ID , Not required
df.drop(['User_ID'],axis=1,inplace=True)
df.head()
```

```python

df['Gender']=pd.get_dummies(df['Gender'],drop_first=1)
df['Gender']
```

```python
# # GENDER IS A CATEGORICAL VARIABLES, CONVERTING CATEGORICAL INTO NUMERICAL - MALE AND FEMALE WE CAN USE 0, 1 HAndling categorical feature Gender - MAPPING MALE AND FEMALE TO 0, 1
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
## Handle categorical feature Age -> See how many unique values are there
df['Age'].unique()
```

```python
#pd.get_dummies(df['Age'],drop_first=True)
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
##fixing categorical City_categort
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df_city.head()
```

```python
# Concat City Category
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
##drop City Category Feature- Which is not required
df.drop('City_Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
## Missing Values
df.isnull().sum()
```

```python
## Focus on replacing missing values
df['Product_Category_2'].unique()
```

```python
## Focus on replacing missing values for Product Category 2 and 3 . The below is a discree feature.. So we need to replace the nan values, We need to use mode, Bast way to replace the missing value is use mode
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].value_counts()
```

```python
# Replace the missing values with mode
df['Product_Category_2'].mode()[0]
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
# Check if it null
df['Product_Category_2'].isnull().sum()
```

```python
### Replace missing values for Product_Category2
## Product_category 3 replace missing values
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```

```python
## Replace the missing values with mode
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```
